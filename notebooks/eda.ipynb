# 30-day readmission 

```python
admission['ADMITTIME'] = pd.to_datetime(admission['ADMITTIME'])
admission['DISCHTIME'] = pd.to_datetime(admission['DISCHTIME'])
admission = admission.sort_values(['SUBJECT_ID', 'ADMITTIME'])
admission['NEXT_ADMITTIME'] = admission.groupby('SUBJECT_ID')['ADMITTIME'].shift(-1)
admission['DAYS_TO_READMIT'] = (admission['NEXT_ADMITTIME']-admission['DISCHTIME']).dt.days
admission['READMIT_30'] = (admission['DAYS_TO_READMIT'] <=30).astype(int)
```

```python
print(admission['READMIT_30'].value_counts(normalize=True)*100)
print(f"\nReadmission rate: {admission['READMIT_30'].mean()*100:.1f}%")
```

```python
admission['READMIT_30'].value_counts().plot(kind='bar')
plt.title('Label Balance: 0 = No Readmission  1 = Readmission within 30 days')
plt.xticks([0,1], ['Not Readmitted', 'Readmitted'], rotation=0)
plt.show()
```

```python
# Age — merge patients to get DOB
merged = admission.merge(patients[['SUBJECT_ID','DOB']], on='SUBJECT_ID')
merged['DOB'] = pd.to_datetime(merged['DOB'])
merged['AGE'] = (merged['ADMITTIME'] - merged['DOB']).dt.days // 365
merged['AGE'].clip(0, 100).hist(bins=20)
plt.title('Patient Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Number of Admissions')
plt.show()
```

```python
admission.merge(patients[['SUBJECT_ID','GENDER']], on='SUBJECT_ID')['GENDER']\
    .value_counts()\
    .plot(kind='bar', color=['steelblue','orange'], edgecolor='white')

plt.title('Patient Gender Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
admission.groupby('ADMISSION_TYPE')['READMIT_30'].mean().plot(kind='bar')
plt.title('Readmission Rate by Admission Type')
plt.ylabel('Rate (0-1)')
plt.show()
```
